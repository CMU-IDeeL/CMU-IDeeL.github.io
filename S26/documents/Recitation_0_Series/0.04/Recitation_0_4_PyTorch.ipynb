## **6. Device**

When you create a tensor, PyTorch stores it on the CPU by default. A `device` is just where a tensor is stored and computed on.

In PyTorch, `device` is usually:
*   CPU → "cpu"
*   GPU → "cuda:0", "cuda:1" etc




```python
x = torch.randn(2, 3)
print("x.device:", x.device)
```

```python
print("Is CUDA available? ", torch.cuda.is_available())
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

x = x.to(device)
print("x is on:", x.device)
```

Note: A model and its input tensors must be on the same device. If your model is on the GPU and your data is on the CPU, you will get an error.

## **7. Vectorization**

Through the exercises in this section, we will begin introducing you to how to think more mathematically through the use of PyTorch by starting with a process known as vectorization.

Index chasing is a very valuable skill, and certainly one you will need in this course, but mathematical problems often have simpler and more efficient representations that use vectors. The process of converting from an implimentation that uses indicies to one that uses vectors is known as vectorization. Once vectorized, the resulting implementation often yields to the user faster and more readable code than before.

$$

\begin{cases}
1, & t \ge 0 \\
0, & t < 0
\end{cases}
$$

$$
X =

x_{m1} & \cdots & x_{mn}

\in \mathbb{R}^{m \times n}

## **9. Automatic Differentiation (Autograd) Basics**

Automatic differentiation is a technique where the computer automatically calculates derivatives (gradients) of our functions for us.

### 9.1. Calculating gradients

Let's take the equation: $
y = x^2
$

Differentiating $y$ w.r.t $x$,
$
\frac{dy}{dx} = 2x
$

At $x = 2$, we have $\frac{dy}{dx} = 2\cdot 2 = 4$.

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2
y.backward()

print(f"dy/dx (x.grad): {x.grad}")
```

`requires_grad=True` tells PyTorch to track operations on that tensor so it can compute and store gradients.

Let's take another example: $
y = a^2 + c\,b.
$

Then
$
\frac{\partial y}{\partial a} = 2a\qquad
\frac{\partial y}{\partial c} = b\qquad
\frac{\partial y}{\partial b} = c
$


At $a = 2$, $b = 4$, $c = 3$, the gradients are:

$
\frac{\partial y}{\partial a} = 2a = 2 \cdot 2 = 4
$

$
\frac{\partial y}{\partial c} = b = 4
$

$
\frac{\partial y}{\partial b} = c = 3
$

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(4.0) # Notice that we are not setting requires_grad to True
c = torch.tensor(3.0, requires_grad=True)

y = a**2 + c * b
y.backward()

print("∂y/∂a (a.grad):", a.grad)
print("∂y/∂c (c.grad):", c.grad)
print("∂y/∂b (b.grad):", b.grad)
```

In the above code cell we set `a.requires_grad = True`, `b.requires_grad = False` (default), and `c.requires_grad = True`,  
so autograd only keeps the gradients
$
\frac{\partial y}{\partial a} = 2a
$ and $
\frac{\partial y}{\partial c} = b,
$
and does not store a gradient for $b$.

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(4.0, requires_grad=True) # Notice that now we are setting requires_grad to True and we will get the gradient for this tensor
c = torch.tensor(3.0, requires_grad=True)

y = a**2 + c * b
y.backward()

print("∂y/∂a (a.grad):", a.grad)
print("∂y/∂c (c.grad):", c.grad)
print("∂y/∂b (b.grad):", b.grad)
```

### 9.2. Computation Graph

PyTorch builds a graph (called computation graph) of all the tensor operations that led to the final tensor (like `y`).
`grad_fn` links that tensor to the last operation that created it.
By walking this graph backwards, PyTorch automatically computes gradeint for all tensors with `requires_grad=True` - Makes life so easier!

image.png

```python
print("y.grad_fn:", y.grad_fn)
print("y.grad_fn.next_functions:")
for nf in y.grad_fn.next_functions:
    print(f"  {nf}")
```

### 9.3. Caution: Gradients accumulate!

```python
x = torch.tensor(2.0, requires_grad=True)

for step in range(4):
    y = x**2
    y.backward() # dy/dx = 2x = 4; added to x.grad
    print(f"Step {step+1}: dy/dx (x.grad) = {x.grad.item()}")
```

By default, `.backward()` accumulates gradients into `x.grad` (or any tensor) instead of replacing them. So, in the above code cell, each call to `y.backward()` computes $4$ and adds it to the existing `x.grad`.


To avoid gradient getting accumulated, we clear the gradient first so that each step uses only the gradient from the current computation, not leftovers from previous steps.

So, we will reset `x.grad` to $0$. before each `.backward()`.

```python
x = torch.tensor(2.0, requires_grad=True)

for step in range(4):
  if x.grad is not None:
    x.grad.zero_() # Resets the gradient to 0
  y = x**2
  y.backward() # Then .backward() computes 4 and adds it to 0
  print(f"Step {step+1}: dy/dx (x.grad) = {x.grad.item()}")
```

## **10. Building and training a Neural Network**

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
```

Now let's convert what we saw in slides to code using the functionalities of PyTorch. By the end of this section, you'll understand the basics of dataloaders/datasets, how to build simple neural networks, and train a model.

We will build a simple fully connected (feed-forward) neural network with three linear layers: one that takes the 28 x 28 pixel image (flattened to 784 features), two hidden layers with 128 and 64 neurons using ReLU activation, and an output layer with 10 neurons (one for each digit class 0 - 9).

image.png

### 10.1. Preparing the data

We will use the MNIST dataset, which contains grayscale images of handwritten digits (0 - 9), each of size 28 x 28 pixels. It has 60,000 training images and 10,000 test images

MNIST_dataset_example.png

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=32, shuffle=True)
testloader = torch.utils.data.DataLoader(testset, batch_size=32, shuffle=False)
```

### 10.2. Building a Neural Network

```python
class SimpleNN(nn.Module):
  def __init__(self):
    super(SimpleNN, self).__init__()
    self.flatten = nn.Flatten()
    self.fc1 = nn.Linear(28*28, 128)
    self.fc2 = nn.Linear(128,64)
    self.fc3 = nn.Linear(64,10)

  def forward(self, x):
    x = self.flatten(x)
    x = self.fc1(x)
    x = F.relu(x)
    x = self.fc2(x)
    x = F.relu(x)
    x = self.fc3(x)
    return x

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleNN().to(device)
```

### 10.3. Defining loss function and optimizer

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

### 10.4. Training the network

```python
print('Starting Training')

for epoch in range(2): # Loop over the dataset multiple time (two times here)
  model.train() # Puts the model in training mode
  running_loss = 0.0
  for i, (inputs, labels) in enumerate(trainloader, start=1): # Go through all the batches of the dataset
    inputs, labels = inputs.to(device), labels.to(device) # All the data and model should be on the same device
    pred = model(inputs) # Getting model predictions
    loss = criterion(pred, labels) # Calculating the loss
    optimizer.zero_grad() # Clearing the gradients to avoid accumulation because we are looping
    loss.backward() # Calculating the gradients
    optimizer.step() # Telling optimizer to update the model's weights

    running_loss += loss.item()
    if i % 100 == 0: # Print the loss every 100 mini-batches
      print(f'[Epoch {epoch+1}, Batch {i}] Training Loss: {running_loss/100:.3f}')
      running_loss = 0.0

print('Finished Training')
```

### 10.5. Test the trained neural network

```python
model.eval() # Puts the model in evaluation/inference mode
correct = 0
total = 0

with torch.no_grad(): # Disables gradient tracking
    for images, labels in testloader:
        images, labels  = images.to(device), labels.to(device)
        outputs   = model(images)
        predicted = outputs.argmax(1)
        correct += (predicted == labels).sum().item()
        total   += labels.numel()

accuracy = 100 * correct / total
print(f'Accuracy on test images: {accuracy:.2f}%')
```
