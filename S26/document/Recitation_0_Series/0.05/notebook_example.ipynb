# Advantages of Notebooks

Notebooks allow you to **mix code and explanations** in one place.

## Why they are useful:
- Run code in small chunks (cells)
- Add text using *Markdown*
- Display images, tables, and plots
- Great for ML experimentation
- Easy to share and reproduce

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.plot(x, y)
plt.title("Sine Wave Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
