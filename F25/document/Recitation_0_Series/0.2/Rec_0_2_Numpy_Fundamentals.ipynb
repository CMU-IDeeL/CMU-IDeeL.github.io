##**1. What is NumPy?**

NumPy is the fundamental package for scientific computing in Python.

NumPy is the most essential library needed for any form of computing dealing with data stored in arrays. In addition to this, NumPy has a wide variety of core submodules that aid in data processing such as `linalg` which supports operations like `norm` or `inv` for matrix inversion.

### Indexing a matrix and drawing a distinction between ij indexing for matrices and xy indexing for images

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 matrix with row-major values for clarity
matrix = np.array([
    [ 0,  1,  2,  3,  4],
    [10, 11, 12, 13, 14],
    [20, 21, 22, 23, 24],
    [30, 31, 32, 33, 34],
    [40, 41, 42, 43, 44],
])

# Access the element at row=3, col=1 using matrix indexing
i, j = 3, 1
val_ij = matrix[i, j]
print(f"Matrix indexing (ij): matrix[3, 1] = {val_ij}")

# Image-style indexing: x is column, y is row
x, y = 1, 3
val_xy = matrix[y, x]
print(f"Image-style indexing (xy): matrix[y=3, x=1] = {val_xy}")

# Display the matrix as an image
plt.figure(figsize=(5, 5))
plt.imshow(matrix, cmap="Blues", origin="upper")
plt.title("Matrix as Image (imshow with origin='upper')")
plt.xlabel("x (column index)")
plt.ylabel("y (row index)")

# Highlight the accessed element
plt.scatter([x], [y], color="red", s=100)
plt.text(x + 0.3, y, f"({x}, {y}) = {matrix[y, x]}", color="black", fontsize=12)

# Draw grid
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, str(matrix[i, j]), ha='center', va='center', color='black')

plt.grid(False)
plt.show()

# Final takeaway print
print("\nSummary:")
print(" - Matrix-style: matrix[i, j]  row i, column j")
print(" - Image-style:  x = column, y = row  matrix[y, x]")
```

## Concatenating matrices along 'dimensions' in NumPy

**Dot Product**

One of the most common NumPy operations well use in machine learning is matrix multiplication using the dot product. Suppose we wanted to take the dot product of two matrices with shapes [2 X 3] and [3 X 2]. We take the rows of our first matrix (2) and the columns of our second matrix (2) to determine the dot product, giving us an output of [2 X 2]. The only requirement is that the inside dimensions match, in this case the first matrix has 3 columns and the second matrix has 3 rows.


dot.gif

```python
# Dot product
a = np.array([[1,2,3], [4,5,6]], dtype=np.float64) # we can specify dtype
b = np.array([[7,8], [9,10], [11, 12]], dtype=np.float64)
c = a.dot(b)
print (f"{a.shape}  {b.shape} = {c.shape}")
print (c)
```

### **b. Slicing: Accessing subsections of Numpy Arrays based on Indices**

```python
# slice along a batch
print(n[0, :, :]) # same as: n[0]

print(n[0, 0:3, 0:4])

# slice along multiple batches
print(n[:, 3, 4]) # returns the elements in the 4th row and 5th column across all batches
```

### **c. Masking and Boolean Indexing**

Masking lets us pick elements that satisfy a logical condition instead of enumerating positions explicitly.  
A Boolean array **mask** with the same shape as the data array acts like a stencil: `masked = data[mask]`.

Typical usecases include removing outliers, selecting classes, or extracting pixels in a regionofinterest.

```python
import numpy as np

arr = np.arange(10)
mask = arr % 2 == 0          # keep even numbers
print("Array:", arr)
print("Mask:", mask)
print(np.unique(mask, return_counts=True))
print("Even elements:", arr[mask], "\n")

# 2D example: keep pixels above a threshold
image = np.arange(16).reshape(4, 4)
high_vals = image > 10
print("Image:\n", image)
print("Pixels > 10:", image[high_vals])
```

### **Squeezing & Expanding Arrays**

#### Squeezing Numpy Arrays

The squeeze operation allows reduction of numpy arrays axes by dropping a specified axis, so long as it is of **unit length**. The product of the shape (overall size of the array) remains the same.

```python
sq = np.arange(4 * 5).reshape(4,-1)
# print(f"Original Array: \n{sq}\n")
print(f"Shape of Original Array: {sq.shape}")

sq_newaxis = sq[:, np.newaxis] # also look at np.expand_dims
print(f"Shape of New Axis Array: {sq_newaxis.shape}")

# transforms from (4, 5, 1) -> (4, 5)
sq1 = np.squeeze(sq_newaxis)
print(f"Shape of Squeezed Array: {sq1.shape}")
```

## **8. Combining data**

### **a. Concatenation**

A concatenation operation joins a sequence of arrays along an *existing* axis. All arrays must either have the same shape (except in the concatenating dimension) or be empty.

```python
a = np.zeros((3, 2, 2))
b = np.zeros((3, 2, 2))

print("Input shapes: a.shape =", a.shape, ", b.shape =", b.shape)

# 2) Concatenate along each axis
for axis in (0, 1, 2):
    c = np.concatenate((a, b), axis=axis)
    print(f"concatenate(axis={axis})  shape {c.shape}")
```

### **b. Stacking**

The stack operation joins a sequence of arrays along a *new* axis. The axis parameter specifies the index of the new axis in the dimensions of the result. For example, if axis = 0 it will be the first dimension and if axis = -1 it will be the last dimension. All arrays need to be of the same size.  The stacked array has one more dimension than the input arrays.

```python
# Stacking 1-d Arrays

array1 = np.array([1, 2, 3])
print("Array 1 is \n", array1, " with dimensions ", array1.shape, "\n")

array2 = np.array([4, 5, 6])
print("Array 2 is \n", array2, " with dimensions ", array2.shape, "\n")


stacked_array1 = np.stack((array1, array2), axis = 0)
print("Stacked array 1 is \n", stacked_array1, " with dimensions ", stacked_array1.shape)

stacked_array2 = np.stack((array1, array2), axis = 1)
print("Stacked array 2 is \n", stacked_array2, " with dimensions ", stacked_array2.shape)

stacked_array3 = np.stack((array1, array2), axis = -1)
print("Stacked array 3 is \n", stacked_array3, " with dimensions ", stacked_array3.shape)
```

### **a. Broadcasting**

What happens when we try to do operations with tensors with seemingly incompatible shapes? Their dimensions arent compatible as is but how does NumPy still gives us the right result? This is where broadcasting comes in. The scalar is broadcast across the vector so that they have compatible shapes.

broadcast.png

```python
# Broadcasting between arrays of different shapes (row vs column)

row_arr = np.array([[1, 2, 3]])
print("Row array:\n", row_arr, "shape:", row_arr.shape, "\n")

col_arr = np.array([[10], [20], [30], [40]])
print("Column array:\n", col_arr, "shape:", col_arr.shape, "\n")

add_arr = row_arr + col_arr
print("row_arr + col_arr =")
print(add_arr, "shape:", add_arr.shape, "\n")

mul_arr = row_arr * col_arr
print("row_arr * col_arr =")
print(mul_arr, "shape:", mul_arr.shape, "\n")
```

### **b. Element-wise operations**

```python
# Element-wise operations on equally shaped arrays

arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("arr1:\n", arr1, "\n")

arr2 = np.array([[10, 20, 30],
                 [40, 50, 60]])
print("arr2:\n", arr2, "\n")

scalar = 5

# Addition with scalar
print("arr1 + 5 =\n", arr1 + scalar, "\n")

# Multiplication with scalar
print("arr1 * 5 =\n", arr1 * scalar, "\n")

# Elementwise addition
print("arr1 + arr2 =\n", arr1 + arr2, "\n")

# Elementwise multiplication (Hadamard product)
print("arr1 * arr2 =\n", arr1 * arr2, "\n")

# Absolute value
print("|10 * arr1| =\n", np.abs(-10 * arr1), "\n")

# Square root
print("sqrt(arr1) =\n", np.sqrt(arr1))
```

**Perils of broadcasting and using the incorrect matrix multiplication operator** Swapping matrix multiplication (@) with element wise multiplication (*)**


Consider a scenario where you want to perform a matrix multiplication. However, by mistake you use an asterisk * instead of a matrix multiplication operator (@). Despite the incorrect operator, the code will still run without error due to the broadcasting functionality in Python. The shapes matching might lead you to think that what you are doing is right whereas in reality, a matrix multiplication was required. You will get wrong values.

```python

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("A shape:", A.shape)
print("B shape:", B.shape)

# Correct matrix multiplication (linearalgebraic dot)
C_mat = A @ B
print("\nA @ B =\n", C_mat)
print("shape:", C_mat.shape)

# Wrong: * does element-wise multiply (with broadcasting)
C_elem = A * B
print("\nA * B =\n", C_elem)
print("shape:", C_elem.shape)
```

Notice:

Both C_mat and C_elem are 22, so if you only check .shape, youd think both are valid.

But

A @ B does the row-by-column sums you expect from linear algebra.

A * B just multiplies each entry of A by the corresponding entry of B.
